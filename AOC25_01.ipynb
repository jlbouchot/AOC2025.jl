```julia
fname = "01_1.txt"
```

```julia


nb0s = open(fname) do f
    pos = 50
    nb0s = 0
    for l in eachline(f)
        if l[1] == 'R'
            pos += parse(Int, l[2:end])
        else
            pos -= parse(Int, l[2:end])
        end
        pos = pos % 100
        if pos == 0 
            nb0s+=1
        end
    end
    (nb0s)
end
     
```

```julia
nb0s = open(fname) do f
    pos = 50
    nb0s = 0
    for l in eachline(f)
        val = parse(Int, l[2:end])
        if val >= 100
            nb0s += div(val,100)
            val = val % 100
        end
        if l[1] == 'R'
            println("Adding ", val, " to ", pos)
            if val + pos >= 100
                nb0s += (pos != 100 ? 1 : 0)
                pos += (val - 100)
            else
                pos += val
            end
            
        else
            println("Removing ", val, " from ", pos)
            if val >= pos
                nb0s += (pos != 0 ? 1 : 0)
                pos -= (val-100)
            else
                pos -= val
            end
            
        end
        println("Pos is now ", pos, " and nb0s is ", nb0s)
    end
    (nb0s)
end
```
