```julia
function find_invalid(n,lrange,rrange)
    start_v = max(10^(n-1), lrange)
    stop_v = min(10^n-1, rrange)
    mod_val = 10^(div(n,2)) + 1
    #println("Multiple of ", mod_val)
    inv = filter(t -> (t % mod_val) == 0, start_v:stop_v)
    #println(inv)
    return sum(inv)
end
```

```julia
function tot_invalid(fname)
    out = open(fname) do f 
        tot_inv = 0
        l = readline(f)
        println(l)
        for r in split(l, ',')
            println(r)
            (lrange, rrange) = parse.(Int,split(r,'-'))
            min_nb_digits = Int(ceil(log10(lrange)))
            max_nb_digits = Int(ceil(log10(rrange)))
            #println("Range (", lrange, ", ", rrange, ") has min digits ", min_nb_digits, " and max digits ", max_nb_digits)
            for n in min_nb_digits:max_nb_digits
                if rem(n,2) == 0
                    tot_inv += find_invalid(n,lrange,rrange)
                end
            end
        end
        (tot_inv)
    end
    return out
end
```

```julia
function find_invalid2(n,lrange,rrange)
    start_v = max(10^(n-1), lrange)
    stop_v = min(10^n-1, rrange)
    out = []
    for length_rep_pattern in 1:n//2
        base = 10^length_rep_pattern
        # println("Base is ", base, " numerator is ", base^(n//length_rep_pattern) -1, " denominator is ", base-1)
        mod_val = (base^(n//length_rep_pattern) -1)/(base-1)
        #println("Multiple of ", mod_val)
        inv = filter(t -> (t % mod_val) == 0, start_v:stop_v)
        #println(inv)
        append!(out, inv)
    end
    
    return (length(out) > 0 ? sum(unique(out)) : 0)
end
```

```julia
function tot_invalid2(fname)
    out = open(fname) do f 
        tot_inv = 0
        l = readline(f)
        for r in split(l, ',')
            (lrange, rrange) = parse.(Int,split(r,'-'))
            min_nb_digits = Int(ceil(log10(lrange)))
            max_nb_digits = Int(ceil(log10(rrange)))
            #println("Range (", lrange, ", ", rrange, ") has min digits ", min_nb_digits, " and max digits ", max_nb_digits)
            for n in min_nb_digits:max_nb_digits
                tot_inv += find_invalid2(n,lrange,rrange)
            end
        end
        (tot_inv)
    end
    return out
end
```
