```julia
function get_max_jolt(fname)
    out = open(fname) do f
        # batteries = []
        tot_joltage = 0
        for l in eachline(f)
            cur_bank = []
            idx = 1
            max_joltage = 0
            idx_max = 0
            length_bank = length(l)
            for c in l
                joltage = parse(Int, c)
                if ((idx < length_bank) && (max_joltage < joltage))
                    max_joltage = joltage
                    idx_max = idx
                end
                idx += 1
                append!(cur_bank, joltage)
            end
            scd_joltage = 0
            max_joltage = max_joltage*10
            for idx in range(idx_max+1, length_bank)
                if (scd_joltage < cur_bank[idx])
                    scd_joltage = cur_bank[idx]
                end
            end
            tot_joltage += max_joltage + scd_joltage
            # push!(batteries, cur_bank)
        end
        # println(batteries)
        (tot_joltage)
    end
    return out
end
```

```julia
function get_max_jolt_rec(vals, l, mem)
    out = 0
    lvals = length(vals)
    if (l == lvals)
        if (mem[l][l] > 0)
            out = mem[l][l]
        else
            for i in range(1,l)
                out = out*10+vals[i]
            end
            mem[l][l] = out
        end
    elseif (l == 1)
        out = maximum(vals)
    else
        if mem[lvals][l] > 0
            out = mem[lvals][l]
        else
            v1 = vals[1]*10^(l-1) + get_max_jolt_rec(vals[2:end], l-1, mem)
            v2 = get_max_jolt_rec(vals[2:end], l, mem)
            out = max(v1, v2)
            mem[lvals][l] = out
        end
    end
    return out
end
```

```julia
l = 7
a = [1,2,3,4,5,6,3,2,1]
get_max_jolt_rec(a, l, [[0 for i in 1:l] for j in 1:length(a)])
```

```julia
function get_max_jolt12(fname)
    length_joltage = 12
    out = open(fname) do f
        # batteries = []
        tot_joltage = 0
        for l in eachline(f)
            cur_bank = []
            for c in l
                joltage = parse(Int, c)
                append!(cur_bank, joltage)
            end
            # push!(batteries, cur_bank)
            mem = [[0 for i in range(1,length_joltage)] for j in range(1,length(l))]
            
            tot_joltage += get_max_jolt_rec(cur_bank,length_joltage, mem)
        end
        # println(batteries)
        
        (tot_joltage)
    end
    return out
end
```
