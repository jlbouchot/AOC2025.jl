```julia
using DataStructures
```

```julia
function parse_file(fname)
    unsorted_intervals = []
    ids = []
    f = open(fname)
    l = readline(f)
    while(l != "")
        push!(unsorted_intervals, parse.(Int, split(l, '-')))
        l = readline(f)
    end
    # Sort by first value, then fuse the intervals if needed
    sort!(unsorted_intervals, by = x -> x[1])
    intervals = [unsorted_intervals[1]]
    for (l,r) in unsorted_intervals[2:end]
        if l <= intervals[end][2]
            intervals[end][2] = max(r, intervals[end][2])
        else
            push!(intervals, [l,r])
        end
    end
    # println(intervals)
    for l in eachline(f)
        append!(ids, parse(Int, l))
    end
    sort!(ids)
    close(f)
    return intervals, ids
end
```

```julia
fname = "05_test.txt"
parse_file(fname)
```

```julia
function how_many_fresh(fname)
    out = 0
    intervals, ids = parse_file(fname)
    idx_interval = 1
    nb_intervals = length(intervals)
    println(nb_intervals, " intervals found")
    for id in ids
        #println(id)
        while (idx_interval <= nb_intervals) && (intervals[idx_interval][2] < id)
            idx_interval += 1
        end
        if idx_interval > nb_intervals
            break
        end
        if (intervals[idx_interval][1] <= id)
            out += 1
            # println("Found id ", id, " in ", intervals[idx_interval])
        end
    end
    return out
end
```

```julia
how_many_fresh(fname)
```

```julia
function how_many_fresh2(fname)
    out = 0
    intervals, ids = parse_file(fname)
    for (l,r) in intervals
        out += (r-l+1)
    end
    return out
end
```
